The data samples are already shuffled.

*Hint: There is a helpful `Data` method for paring down a dataset into a certain range of samples.*

```python
# Fit set
polyData = data.Data(filepath = 'data/poly_data.csv')
numSamp = polyData.get_num_samples()
fitSet = polyData.limit_samples(0, numSamp // 2)
fitSet = data.Data(filepath = 'data/poly_data.csv', headers=['X', 'Y'], data=fitSet, header2col={'X': 0, 'Y': 1})
print('fit set: \n', fitSet)
```

```python
# Validation set
polyData = data.Data(filepath='data/poly_data.csv')
numSamples = polyData.get_num_samples()
validationSet = polyData.limit_samples(numSamples // 2, numSamples)
validationSet = data.Data(filepath='data/poly_data.csv', headers=['X', 'Y'], data=validationSet, header2col={'X': 0, 'Y': 1})
print(print('validation set: \n', validationSet))
```
